```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

miami_qb_df = pd.read_csv('miami_qb_data.csv')
miami_qb_df.columns = miami_qb_df.columns.str.strip()
miami_qb_df['Rating'] = pd.to_numeric(miami_qb_df['Rating'], errors='coerce')

miami_rating_yearly = miami_qb_df.groupby('Year')['Rating'].mean().reset_index()

cleaned_df = pd.read_csv('Data/qb_data_cleaned.csv')
cleaned_df.columns = cleaned_df.columns.str.strip()
num_cols = ['Cmp', 'Att', 'Yds', 'TD', 'Int']
for col in num_cols:
    cleaned_df[col] = pd.to_numeric(cleaned_df[col], errors='coerce')

def nfl_passer_rating(row):
    try:
        a = ((row['Cmp'] / row['Att']) - 0.3) * 5
        b = ((row['Yds'] / row['Att']) - 3) * 0.25
        c = (row['TD'] / row['Att']) * 20
        d = 2.375 - ((row['Int'] / row['Att']) * 25)
        rating = ((max(0, min(a, 2.375)) + max(0, min(b, 2.375)) + max(0, min(c, 2.375)) + max(0, min(d, 2.375))) / 6) * 100
        return rating
    except:
        return np.nan

team_year_stats = cleaned_df[cleaned_df['FantPos'] == 'QB'].groupby(['Tm', 'Year']).agg({
    'Cmp': 'sum',
    'Att': 'sum',
    'Yds': 'sum',
    'TD': 'sum',
    'Int': 'sum'
}).reset_index()
team_year_stats['Rating'] = team_year_stats.apply(nfl_passer_rating, axis=1)

league_team_rating = team_year_stats[team_year_stats['Tm'] != 'MIA'].groupby('Year')['Rating'].mean().reset_index()

rating_compare = pd.merge(miami_rating_yearly, league_team_rating, on='Year', how='inner', suffixes=('_MIA', '_League'))

plt.figure(figsize=(12,6))
plt.plot(rating_compare['Year'], rating_compare['Rating_MIA'], marker='o', color='#008E97', label='Dolphins Official Rating', linewidth=2)
plt.plot(rating_compare['Year'], rating_compare['Rating_League'], marker='o', color='gray', linestyle='--', label='League Avg Team Rating', linewidth=2)
plt.title('Dolphins QB Passer Rating vs. League Avg Team Rating (Yearly)')
plt.xlabel('Year')
plt.ylabel('Passer Rating')
plt.xticks(rating_compare['Year'], rotation=45)
plt.ylim(60, 110)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np

cleaned_df = pd.read_csv('Data/qb_data_cleaned.csv')
cleaned_df.columns = cleaned_df.columns.str.strip()
qb_df = cleaned_df[cleaned_df['FantPos'] == 'QB'].copy()
for col in ['Yds', 'TD', 'Int', 'Cmp', 'Att']:
    qb_df[col] = pd.to_numeric(qb_df[col], errors='coerce')

all_years = list(range(2004, 2025))
qb_df = qb_df[(qb_df['Year'] >= 2004) & (qb_df['Year'] <= 2024)]

mia_qb = qb_df[qb_df['Tm'] == 'MIA'].copy()
mia_yearly = mia_qb.groupby('Year').agg({
    'Yds': 'sum',
    'TD': 'sum',
    'Int': 'sum',
    'Cmp': 'sum',
    'Att': 'sum',
}).reindex(all_years).reset_index()
mia_yearly['CmpPct'] = (mia_yearly['Cmp'] / mia_yearly['Att']) * 100
mia_yearly['TD/INT'] = mia_yearly['TD'] / mia_yearly['Int'].replace(0, np.nan)

other_teams = qb_df[qb_df['Tm'] != 'MIA']
other_team_sums = other_teams.groupby(['Year','Tm']).agg({
    'Yds': 'sum',
    'TD': 'sum',
    'Int': 'sum',
    'Cmp': 'sum',
    'Att': 'sum',
}).reset_index()
other_team_sums['CmpPct'] = (other_team_sums['Cmp'] / other_team_sums['Att']) * 100
league_yearly = other_team_sums.groupby('Year').agg({
    'Yds': 'mean',
    'TD': 'mean',
    'Int': 'mean',
    'Cmp': 'mean',
    'Att': 'mean',
    'CmpPct': 'mean',
}).reindex(all_years).reset_index()
league_yearly['TD/INT'] = league_yearly['TD'] / league_yearly['Int'].replace(0, np.nan)

try:
    mia_qb_data = pd.read_csv('miami_qb_data.csv')
    mia_qb_data.columns = mia_qb_data.columns.str.strip()
    mia_qb_data['Rating'] = pd.to_numeric(mia_qb_data['Rating'], errors='coerce')
    mia_rating = mia_qb_data.groupby('Year')['Rating'].mean().reset_index()
    mia_yearly = pd.merge(mia_yearly, mia_rating, on='Year', how='left')
except Exception as e:
    print('Could not load Dolphins official rating:', e)

compare = pd.merge(mia_yearly, league_yearly, on='Year', suffixes=('_MIA', '_League'))

def get_percentile(stat, year, value):
    year_stats = qb_df[qb_df['Year'] == year].groupby('Tm').agg({'Cmp': 'sum', 'Att': 'sum', 'Yds': 'sum', 'TD': 'sum', 'Int': 'sum'})
    year_stats['CmpPct'] = (year_stats['Cmp'] / year_stats['Att']) * 100
    if stat == 'CmpPct':
        stat_series = year_stats['CmpPct']
    else:
        stat_series = year_stats[stat]
    if stat == 'Int':
        pct = 1 - (stat_series < value).mean()
    else:
        pct = (stat_series < value).mean()
    return pct * 100

percentiles = []
for _, row in mia_yearly.iterrows():
    year = row['Year']
    percentiles.append({
        'Year': year,
        'Yds': get_percentile('Yds', year, row['Yds']),
        'TD': get_percentile('TD', year, row['TD']),
        'Int': get_percentile('Int', year, row['Int']),
        'CmpPct': get_percentile('CmpPct', year, row['CmpPct'])
    })
percentile_df = pd.DataFrame(percentiles)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

miami_records = pd.DataFrame([
    [2004, 4, 12, 0, 0.250], [2005, 9, 7, 0, 0.563], [2006, 6, 10, 0, 0.375], [2007, 1, 15, 0, 0.063],
    [2008, 11, 5, 0, 0.688], [2009, 7, 9, 0, 0.438], [2010, 7, 9, 0, 0.438], [2011, 6, 10, 0, 0.375],
    [2012, 7, 9, 0, 0.438], [2013, 8, 8, 0, 0.500], [2014, 8, 8, 0, 0.500], [2015, 6, 10, 0, 0.375],
    [2016, 10, 6, 0, 0.625], [2017, 6, 10, 0, 0.375], [2018, 7, 9, 0, 0.438], [2019, 5, 11, 0, 0.313],
    [2020, 10, 6, 0, 0.625], [2021, 9, 8, 0, 0.529], [2022, 9, 8, 0, 0.529], [2023, 11, 6, 0, 0.647],
    [2024, 8, 9, 0, 0.471],
], columns=["Year", "Wins", "Losses", "Ties", "WinPct"])
miami_records["Win %"] = (miami_records["WinPct"] * 100).round(1)

cleaned_df = pd.read_csv('Data/qb_data_cleaned.csv')
cleaned_df.columns = cleaned_df.columns.str.strip()
miami_qb = cleaned_df[(cleaned_df['FantPos'] == 'QB') & (cleaned_df['Tm'] == 'MIA')]

miami_qb['TD'] = pd.to_numeric(miami_qb['TD'], errors='coerce')
miami_qb['Int'] = pd.to_numeric(miami_qb['Int'], errors='coerce')
miami_qb['Yds'] = pd.to_numeric(miami_qb['Yds'], errors='coerce')
miami_qb['Cmp'] = pd.to_numeric(miami_qb['Cmp'], errors='coerce')
miami_qb['Att'] = pd.to_numeric(miami_qb['Att'], errors='coerce')
miami_qb['CmpPct'] = miami_qb['Cmp'] / miami_qb['Att'] * 100

agg_stats = miami_qb.groupby('Year').agg({'TD': 'sum', 'Int': 'sum', 'Yds': 'sum', 'CmpPct': 'mean'}).reset_index()

hist_df = pd.merge(miami_records[['Year', 'Win %']], agg_stats, on='Year', how='inner')

stats = ['TD', 'Int', 'Yds', 'CmpPct']
plt.figure(figsize=(16, 10))
for i, stat in enumerate(stats, 1):
    plt.subplot(2, 2, i)
    sns.histplot(hist_df[stat], kde=True, color='#008E97', bins=10)
    plt.xlabel(stat)
    plt.title(f'{stat} Distribution (Dolphins QBs)')
    # Show correlation with win percentage
    corr = hist_df['Win %'].corr(hist_df[stat])
    plt.annotate(f'Corr with Win %: {corr:.2f}', xy=(0.05, 0.9), xycoords='axes fraction', fontsize=12, color='black')
plt.tight_layout()
plt.show()

corrs = hist_df.corr(numeric_only=True)['Win %'][stats]
print('Correlation of Dolphins Win % with QB Stats:')
print(corrs.round(2))
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

cleaned_df = pd.read_csv('Data/qb_data_cleaned.csv')
cleaned_df.columns = cleaned_df.columns.str.strip()
qb_df = cleaned_df[cleaned_df['FantPos'] == 'QB'].copy()
for col in ['Yds', 'TD', 'Int', 'Cmp', 'Att']:
    qb_df[col] = pd.to_numeric(qb_df[col], errors='coerce')
qb_df['CmpPct'] = qb_df['Cmp'] / qb_df['Att'] * 100

all_years = list(range(2004, 2025))
qb_df = qb_df[(qb_df['Year'] >= 2004) & (qb_df['Year'] <= 2024)]

mia_team = qb_df[qb_df['Tm'] == 'MIA'].groupby('Year')[['Yds','TD','Int','Cmp','Att']].sum().reindex(all_years).reset_index()
mia_team['CmpPct'] = mia_team['Cmp'] / mia_team['Att'] * 100

other_teams = qb_df[qb_df['Tm'] != 'MIA']
other_team_sums = other_teams.groupby(['Year','Tm'])[['Yds','TD','Int','Cmp','Att']].sum().reset_index()
other_team_avgs = other_team_sums.groupby('Year')[['Yds','TD','Int','Cmp','Att']].mean().reindex(all_years).reset_index()
other_team_avgs['CmpPct'] = other_team_avgs['Cmp'] / other_team_avgs['Att'] * 100

miami_vals = mia_team[['Yds','TD','Int','Cmp','Att','CmpPct']].mean().tolist()
other_vals = other_team_avgs[['Yds','TD','Int','Cmp','Att','CmpPct']].mean().tolist()
stats = ['Yds', 'TD', 'Int', 'Cmp', 'Att', 'CmpPct']

for i, stat in enumerate(stats):
    plt.figure(figsize=(5,5))
    plt.bar(['Miami QBs', 'Other Teams (Avg Team)'], [miami_vals[i], other_vals[i]], color=['#008E97', 'gray'])
    plt.ylabel(stat)
    plt.title(f'{stat}: Miami QBs vs Other Teams (2004-2024 Avg Team Total)')
    for j, val in enumerate([miami_vals[i], other_vals[i]]):
        plt.text(j, val + (max(miami_vals[i], other_vals[i]) * 0.01), f'{val:.1f}', ha='center', va='bottom', fontsize=12)
    plt.tight_layout()
    plt.show()
```
